```python
def year(df):
    # 'pop'으로 끝나는 열만 추출
    pop_cols = [col for col in df.columns if col.endswith('pop')]

    # 필드값(field_1)과 연도(year)별 합계 계산
    grouped = df.groupby(['field_1', 'year'])[pop_cols].sum()
    grouped['pop_total'] = grouped.sum(axis=1)-grouped['total_pop'] # 모든 pop 값의 합계

    # 연도별 pop_total 값을 열로 변환
    averages = grouped['pop_total'].unstack(level=1).reset_index()
    averages.columns = ['field_1'] + [f'{year}총합' for year in averages.columns[1:]]

    # 최종 데이터프레임 생성
    final_result = df[['field_1']].drop_duplicates().merge(averages, on='field_1', how='outer')

    return final_result

# 예제 사용
HS_연총합 = year(HS_공유차량_거주인구)
HN_연총합 = year(HN_공유차량_거주인구)
```

```python
def aggregate_population_field(df):
    # 'pop'으로 끝나는 컬럼 선택
    pop_columns = [col for col in df.columns if col.endswith('pop')]
    aggregated = df.groupby('field_1')[pop_columns].mean().reset_index()

    return aggregated

HS_공유차량_거주인구= aggregate_population_field(HS_공유차량_거주인구)
HN_공유차량_거주인구= aggregate_population_field(HN_공유차량_거주인구)
```

```python
def calculate(pop_gdf):
    pop_cols = [col for col in pop_gdf.columns if 'pop' in col]

    total_pop = pop_gdf[pop_cols].sum(axis=1)-pop_gdf['total_pop']
    male_pop = pop_gdf[[col for col in pop_cols if col.startswith('m')]].sum(axis=1)
    female_pop = pop_gdf[[col for col in pop_cols if col.startswith('w')]].sum(axis=1)
    drivable_cols = [col for col in pop_cols if any(age in col for age in ['20g', '30g', '40g', '50g', '60g', '70g'])]
    high_risk_cols = [col for col in pop_cols if any(age in col for age in ['80g', '90g', '100g'])]

    drivable_pop = pop_gdf[drivable_cols].sum(axis=1)
    high_risk_pop = pop_gdf[high_risk_cols].sum(axis=1)

    calculated_df = pd.DataFrame({
        'field_1' : pop_gdf['field_1'],
        '년 평균 total': total_pop,
        'male': male_pop,
        'female': female_pop,
        'drivable': drivable_pop,
        'high_risk': high_risk_pop
    })

    return calculated_df

HS_공유차량_거주인구= calculate(HS_공유차량_거주인구)
HN_공유차량_거주인구= calculate(HN_공유차량_거주인구)
```

```python
import geopandas as gpd
from shapely.geometry import Point, box

## 임대료 데이터 처리
HS_공공주택_임대료 = Lon_Lat(HS_공공주택_임대료)
HN_공공주택_임대료 = Lon_Lat(HN_공공주택_임대료)

# 3. 임대료(pay_rent): 평균 계산 함수
def calculate_average_rent_within_buffer(gdf, payment_gdf, buffer_size):
    # 각 점의 버퍼 박스 계산
    minx = gdf['geometry'].x - buffer_size
    miny = gdf['geometry'].y - buffer_size
    maxx = gdf['geometry'].x + buffer_size
    maxy = gdf['geometry'].y + buffer_size

    # 각 점의 버퍼 박스를 GeoDataFrame으로 생성
    buffer_boxes = [box(x1, y1, x2, y2) for x1, y1, x2, y2 in zip(minx, miny, maxx, maxy)]
    buffer_gdf = gpd.GeoDataFrame(geometry=buffer_boxes)

    # 각 버퍼에 포함된 임대료 데이터를 확인
    avg_rents = []
    for buffer_box in buffer_gdf['geometry']:
        # 해당 버퍼에 포함된 데이터 필터링
        within_i = payment_gdf[payment_gdf.within(buffer_box)]

        # 임대료 평균 계산
        avg_rent = within_i['pay_rent'].mean() if not within_i.empty else 0
        avg_rents.append(avg_rent)

    return avg_rents

# 4. 데이터 처리: 데이터프레임에 새로운 열 추가
buffer_size = 0.01  # 버퍼 크기 설정

# 임대료(pay_rent) 평균값 계산
HS_공공['avg_rent'] = calculate_average_rent_within_buffer(HS_공공, HS_공공주택_임대료, buffer_size)
HS_행복['avg_rent'] = calculate_average_rent_within_buffer(HS_행복, HS_공공주택_임대료, buffer_size)
HS_영구['avg_rent'] = calculate_average_rent_within_buffer(HS_영구, HS_공공주택_임대료, buffer_size)
HS_국민['avg_rent'] = calculate_average_rent_within_buffer(HS_국민, HS_공공주택_임대료, buffer_size)

HN_공공['avg_rent'] = calculate_average_rent_within_buffer(HN_공공, HN_공공주택_임대료, buffer_size)
HN_행복['avg_rent'] = calculate_average_rent_within_buffer(HN_행복, HN_공공주택_임대료, buffer_size)
HN_영구['avg_rent'] = calculate_average_rent_within_buffer(HN_영구, HN_공공주택_임대료, buffer_size)
HN_국민['avg_rent'] = calculate_average_rent_within_buffer(HN_국민, HN_공공주택_임대료, buffer_size)
```

```python
def calculate_all_pyng_frequencies_multilevel(gdf, payment_gdf, buffer_size=0.01):

    # 각 점의 버퍼 계산
    minx = gdf['geometry'].x - buffer_size
    miny = gdf['geometry'].y - buffer_size
    maxx = gdf['geometry'].x + buffer_size
    maxy = gdf['geometry'].y + buffer_size

    # 버퍼 박스 생성
    buffer_boxes = [box(x1, y1, x2, y2) for x1, y1, x2, y2 in zip(minx, miny, maxx, maxy)]
    buffer_gdf = gpd.GeoDataFrame(geometry=buffer_boxes, index=gdf.index)

    # 빈도 데이터를 저장할 리스트
    frequency_data = []

    # 각 버퍼 내의 데이터 필터링 및 빈도 계산
    for buffer_box in buffer_gdf['geometry']:
        # 버퍼 내 데이터 추출
        within_i = payment_gdf[payment_gdf.within(buffer_box)]

        # 빈도 계산
        if not within_i.empty:
            freq_counts = within_i['pyng'].value_counts().to_dict()
        else:
            freq_counts = {category: 0 for category in payment_gdf['pyng'].unique()}

        frequency_data.append(freq_counts)

    # 빈도 데이터를 정리하여 DataFrame 생성
    categories = sorted(payment_gdf['pyng'].unique())
    frequency_df = pd.DataFrame(frequency_data, index=gdf.index).fillna(0)

    # 멀티 인덱스 설정
    frequency_df = frequency_df.reindex(columns=categories, fill_value=0).astype(int)
    columns = pd.MultiIndex.from_tuples(
        [("평", category) for category in categories],
        names=["상위 인덱스", "구분"]
    )
    frequency_df.columns = columns

    return frequency_df

def merge_with_multilevel_columns(base_df, frequency_df):
    """
    멀티컬럼을 가진 frequency_df를 base_df와 병합.
    """
    base_df = pd.concat([base_df, frequency_df], axis=1)
    base_df.columns = pd.MultiIndex.from_tuples(
        [(col[0], col[1]) if isinstance(col, tuple) else (col, '') for col in base_df.columns],
        names=["상위 인덱스", "구분"]
    )
    return base_df

# 멀티컬럼 유지 병합
공공_df = calculate_all_pyng_frequencies_multilevel(HS_공공, HS_공공주택_임대료, buffer_size)
HS_공공 = merge_with_multilevel_columns(HS_공공, 공공_df)

영구_df = calculate_all_pyng_frequencies_multilevel(HS_영구, HS_공공주택_임대료, buffer_size)
HS_영구 = merge_with_multilevel_columns(HS_영구, 영구_df)

행복_df = calculate_all_pyng_frequencies_multilevel(HS_행복, HS_공공주택_임대료, buffer_size)
HS_행복 = merge_with_multilevel_columns(HS_행복, 행복_df)

국민_df = calculate_all_pyng_frequencies_multilevel(HS_국민, HS_공공주택_임대료, buffer_size)
HS_국민 = merge_with_multilevel_columns(HS_국민, 국민_df)

공공_df = calculate_all_pyng_frequencies_multilevel(HN_공공, HN_공공주택_임대료, buffer_size)
HN_공공 = merge_with_multilevel_columns(HN_공공, 공공_df)

영구_df = calculate_all_pyng_frequencies_multilevel(HN_영구, HN_공공주택_임대료, buffer_size)
HN_영구 = merge_with_multilevel_columns(HN_영구, 영구_df)

행복_df = calculate_all_pyng_frequencies_multilevel(HN_행복, HN_공공주택_임대료, buffer_size)
HN_행복 = merge_with_multilevel_columns(HN_행복, 행복_df)

국민_df = calculate_all_pyng_frequencies_multilevel(HN_국민, HN_공공주택_임대료, buffer_size)
HN_국민 = merge_with_multilevel_columns(HN_국민, 국민_df)
```

# 공유차량 진행 기간 (month) (사용 X)

```python
def month(df):
    # MultiIndex의 첫 번째 레벨에서 이름을 가져와 필터링
    tm_columns = [col for col in df.columns.get_level_values(0) if str(col).startswith('tm')]

    # 결측치가 아닌 값의 개수 계산
    df['months'] = df[tm_columns].notna().sum(axis=1)
    return df
```
